SCATTER PLOT FOR TIPS

```python
scs.scatterplot(x="tip",y="total_bill",data=tips)
```

```python
scs.scatterplot(x="tip",y="total_bill",data=tips,hue="day",size="size",palette="YlGnBu")
```

```python
scs.histplot(tips['tip'], kde= 'true',bins=15)
```

```python
scs.barplot(x="sex", y="tip", data=tips, palette="YlGnBu")
```

```python
scs.boxplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu")
```

```python
scs.stripplot(x="day",y="tip", data=tips, hue="sex", dodge='true', palette="YlGnBu")
```

```python
scs.jointplot(x="tip",y="total_bill", data=tips)
```

```python
scs.jointplot(x="tip",y="total_bill", data=tips, kind="reg",)
```

```python
scs.jointplot(x="tip",y="total_bill", data=tips, kind="kde", shade='true', cmap="YlGnBu")
```

```python
numeric_titanic = titanic.select_dtypes(include=['number'])  # Select only numeric columns
correlation_matrix = numeric_titanic.corr()
print(correlation_matrix)
```

```python
numeric_titanic = numeric_titanic.dropna()
```

```python
numeric_titanic = numeric_titanic.fillna(0)  # Replace NaNs with 0
```

```python
import pandas as pd
for column in numeric_titanic.columns:
    numeric_titanic[column] = pd.to_numeric(numeric_titanic[column], errors='coerce')
```

```python
# Select numeric columns
numeric_titanic = titanic.select_dtypes(include=['number'])

# Handle missing values
numeric_titanic = numeric_titanic.fillna(0)  # Replace NaNs with 0 (or use dropna if preferred)

# Compute correlation matrix
correlation_matrix = numeric_titanic.corr()

# Print correlation matrix
print(correlation_matrix)
```

```python
scs.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")
```

```python
scs.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
```

```python
scs.heatmap(correlation_matrix, annot=True, cmap="icefire")
```
