SIMPLE CALCULATOR

```python
# Calculator
number1 = float(input("Enter the first number: "))
number2 = float(input("Enter the second number: "))

def addition():
    add = number1 + number2
    print("Addition =", add)

def subtraction(): 
    sub = number1 - number2
    print("Subtraction =", sub)

def division(): 
    if number2 != 0:
        div = number1 / number2
        print("Division =", div)
    else:
        print("Division by zero is not allowed.")

def multiply(): 
    mult = number1 * number2
    print("Multiplication =", mult)

def modulus(): 
    if number2 != 0:
        mod = number1 % number2
        print("Modulus =", mod)
    else:
        print("Modulus by zero is not allowed.")

# Call all functions
addition()
subtraction()
division()
multiply()
modulus()
    
```

```python
# Homework:  
# A more comprehensive example combining concepts
import pandas as pd
import numpy as np

# Create student data
student_data = {
    'Student_ID': range(1, 100),
    'Math': np.random.randint(60, 100, 99),
    'Science': np.random.randint(60, 100, 99),
    'Literature': np.random.randint(60, 100, 99)
}

df = pd.DataFrame(student_data)
```

```python
df.head(n=4)
```

```python
df.describe()
```

```python
print("\nMeans")
for col in ['Math','Science','Literature']:
    print(f"{col}:{df[col].mean():.2f}")
print("\nMaximum")
for col in ['Math','Science','Literature']:
    print(f"{col}:{df[col].max():.2f}")
print("\nMinmum")
for col in ['Math','Science','Literature']:
    print(f"{col}:{df[col].min():.2f}")
print("\nStd")
for col in ['Math','Science','Literature']:
    print(f"{col}:{df[col].std():.2f}")
```
